```python
import matplotlib.pyplot as plt
import dartwork_mpl as dm
import numpy as np
from matplotlib.transforms import blended_transform_factory

# Use scientific preset for clean look
dm.style.use_preset("scientific")
```

```python
# Data Definition
labels = [
    "Frontend Development",
    "Backend Architecture",
    "Data Engineering",
    "API Integration",
    "Quality Assurance",
    "DevOps & Infrastructure",
    "Security Compliance",
    "Technical Documentation"
]

# Approximate values for the scenario
# Negative: Review & Refactoring overhead
# Positive: Code Generation savings
neg_values = np.array([-5, -8, -10, -10, -8, -9, -10, -7])
pos_values = np.array([20, 35, 32, 40, 20, 28, 38, 30])

# Calculate Total (Average)
labels.append("Total")
neg_values = np.append(neg_values, np.mean(neg_values))
pos_values = np.append(pos_values, np.mean(pos_values))

# Reverse order for barh (top to bottom)
labels = labels[::-1]
neg_values = neg_values[::-1]
pos_values = pos_values[::-1]
```

```python
# Plotting Logic
fig = plt.figure(figsize=(dm.cm2in(10), dm.cm2in(12)), dpi=300)

# Vertical positioning - cascading from top to bottom
title_y = 0.98  # Starting y-coordinate for title (topmost element)
title_to_legend_gap = 0.05  # Gap between title and legend
legend_y = title_y - title_to_legend_gap  # Legend y-coordinate (auto-calculated)
legend_to_figure_gap = 0.07  # Gap between legend and figure
figure_top = legend_y - legend_to_figure_gap  # Figure top margin (auto-calculated)

# Layout parameters
left_margin = 0.35
right_margin = 0.95
figure_bottom = 0.1
base_x = 0.02  # Common x-coordinate for title, legend, and labels

# 1. Dynamic Layout Calculation
# We need to reserve space on the left for the labels.
# The labels will be drawn at figure x=0.02.
# We need to ensure the axes starts after the longest label.
# Let's keep the left margin generous.
gs = fig.add_gridspec(1, 1, 
                      left=left_margin, 
                      right=right_margin, 
                      top=figure_top, 
                      bottom=figure_bottom)
ax = fig.add_subplot(gs[0, 0])

# 2. Plot Bars
hbar_height = 0.5
hbar_spacing_factor = 1.6  # Spacing between bars as multiple of hbar_height
y_pos = np.arange(len(labels)) * hbar_height * hbar_spacing_factor

# Colors
color_neg = "#191970"  # MidnightBlue-like
color_pos = "#6495ED"  # CornflowerBlue-like

bars_neg = ax.barh(y_pos, neg_values, 
                   height=hbar_height, color=color_neg, 
                   label="Review & Refactoring overhead")
bars_pos = ax.barh(y_pos, pos_values, 
                   height=hbar_height, color=color_pos, 
                   label="Code Generation savings")

# 3. Styling and Labels
# Remove spines
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Remove x and y ticks
ax.set_xticks([])
ax.set_yticks([])

# Add vertical grid line at 0
ax.axvline(0, color='lightgray', linewidth=0.8)

# Transform for labels: x in figure coords, y in data coords
transform = blended_transform_factory(fig.transFigure, ax.transData)

# Add Text Labels (Left Aligned Indexers)
for i, label in enumerate(labels):
    # Bold 'Total' using dm.fw
    weight = dm.fw(1) if label == 'Total' else dm.fw(0)
    
    # Position: x=base_x (figure), y=y_pos[i] (data)
    ax.text(base_x, y_pos[i], label, ha='left', va='center', transform=transform, 
            fontsize=dm.fs(0), fontweight=weight, wrap=True)

# Add Value Labels
for rect in bars_neg:
    width = rect.get_width()
    # Label position: left of the bar
    ax.text(width - 1, rect.get_y() + rect.get_height()/2, f"{int(width)}", 
            ha='right', va='center', fontsize=dm.fs(-1))

for rect in bars_pos:
    width = rect.get_width()
    # Label position: right of the bar
    label_text = f"{int(width)}"
    ax.text(width + 1, rect.get_y() + rect.get_height()/2, label_text,
            ha='left', va='center', 
            fontsize=dm.fs(-1))

# Title and Legend
# Title positioned at top, with cascading layout below
fig.text(base_x, title_y, "Engineering hours shifted by AI assistants, % of sprint capacity", 
         fontsize=dm.fs(2), 
         fontweight=dm.fw(1), 
         ha='left')

# Custom Legend
# Positioned below title, automatically calculated to avoid overlap
fig.legend(loc='upper left', bbox_to_anchor=(base_x, legend_y), 
           ncol=2, 
           frameon=False, 
           fontsize=dm.fs(0), 
           borderaxespad=0,
           columnspacing=1.5)


plt.show()
```
